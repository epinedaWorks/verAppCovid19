## Aplicación para conocer el comportamiento de Covid19.
**La información que se presenta se enfoca especialmente en la región centroamericana incluyendo a México, sin embargo, también se presenta la información del resto de países.**   
Se recomienda ver la aplicación en computadora o tablet, en caso de que sea smartphone se recomienda rotar la pantalla.

Link para acceder:   
 
https://bit.ly/gtcovid19
